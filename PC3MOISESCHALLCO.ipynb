```python
#Pregunta 3 y 4
#definimos el dataframe inicial
import pandas as pd
fallecidos = pd.read_csv('fallecidosCOVID.csv')
fallecidos
```

```python
type(fallecidos)
```

```python
#Separamos en columnas pertinentes la data desordenada
fallecidos['FECHA_CORTE;FECHA_FALLECIMIENTO;EDAD_DECLARADA;SEXO;CLASIFICACION_DEF;DEPARTAMENTO;PROVINCIA;DISTRITO;UBIGEO;UUID'] = fallecidos['FECHA_CORTE;FECHA_FALLECIMIENTO;EDAD_DECLARADA;SEXO;CLASIFICACION_DEF;DEPARTAMENTO;PROVINCIA;DISTRITO;UBIGEO;UUID'].astype(str)
columnas_nuevas = ['FECHA_CORTE', 'FECHA_FALLECIMIENTO', 'EDAD_DECLARADA', 'SEXO', 
                   'CLASIFICACION_DEF', 'DEPARTAMENTO', 'PROVINCIA', 'DISTRITO', 
                   'UBIGEO', 'UUID']
fallecidos[columnas_nuevas] = fallecidos['FECHA_CORTE;FECHA_FALLECIMIENTO;EDAD_DECLARADA;SEXO;CLASIFICACION_DEF;DEPARTAMENTO;PROVINCIA;DISTRITO;UBIGEO;UUID'].str.split(';', expand=True)
fallecidos.drop(columns=['FECHA_CORTE;FECHA_FALLECIMIENTO;EDAD_DECLARADA;SEXO;CLASIFICACION_DEF;DEPARTAMENTO;PROVINCIA;DISTRITO;UBIGEO;UUID'], inplace=True)
print(fallecidos.head())
```

```python
fallecidos
```

```python
fallecidos['FECHA_FALLECIMIENTO'].value_counts()
#Antes de hacer el pivottable vemos las frecuencias de cada fecha exacta
```

```python
#Como los datos de fallecimiento no estaban en formato de fecha se los cambia y se los separa.
fallecidos['FECHA_FALLECIMIENTO'] = pd.to_datetime(fallecidos['FECHA_FALLECIMIENTO'], format='%Y%m%d')
fallecidos['AÑO'] = fallecidos['FECHA_FALLECIMIENTO'].dt.year
fallecidos['MES'] = fallecidos['FECHA_FALLECIMIENTO'].dt.month
fallecidos['DÍA'] = fallecidos['FECHA_FALLECIMIENTO'].dt.day
fallecidos
```

```python
#Se examina la frecuencia de los años
dato1 = fallecidos['AÑO'].value_counts()
dato1
```

```python
#Se examina la frecuencia de los meses
dato2 = fallecidos['MES'].value_counts()
dato2
```

```python
#Se elabora la tabla pedida en la pregunta 3
frecuencia = fallecidos.groupby(['AÑO', 'MES']).size().reset_index(name='FRECUENCIA')
frecuencia
```

```python
#Se elabora la pivot table pedida en (4)
reporte2 = pd.pivot_table(
    fallecidos,
    index=['AÑO', 'MES'],
    columns='CLASIFICACION_DEF',
    aggfunc='size',
    fill_value=0
).reset_index()

reporte2
```
